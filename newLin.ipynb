# Predicting Sales Based On
# Money Spent In TV Marketing
# Using Linear Regression

```python
# importing pandas libray for CSV file I/O
import pandas as pd
# Importing the dataset
df = pd.read_csv("advertising.csv")
```

```python
print('First 5 rows of the dataset: ')
print(df.head())
```

```python
print('Shape of the Data frame: ', df.shape)
```

```python
print("The summary of dataset: ")
print(df.info())
```

```python
print("Check for missing values: ")
print(df.isnull().sum())
```

```python
# There are no missing values in variables
```

```python
# Declaring the feature vector and target variable
X = df[['TV']]
y = df['Sales']
```

```python
# Splitting X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print('Shape of Training set of features: ', X_train.shape)
print('Shape of Testing set of features: ', X_test.shape)
print('Shape of Training set of target: ', y_train.shape)
print('Shape of Testing set of target: ', y_test.shape)
```

```python
# Import Linear Regression Model
from sklearn.linear_model import LinearRegression
# Instantiate the Model
slr= LinearRegression()
# Fitting the model with training set
slr.fit(X_train, y_train)
```

```python
# Calculating the model coefficients
intercept = slr.intercept_
coefficient = slr.coef_
print('Intercept: ', intercept)
print('Coefficient:', coefficient[0])
```

```python
# The Regression Equation
print('Regression Equation: Sales = 7.162 + 0.0544 * TV')
```

```python
# Importing Matplotlib
import matplotlib.pyplot as plt
# Plotting the Line of Best Fit
plt.scatter(X_train, y_train)
plt.plot(X_train, intercept + coefficient * X_train, 'r')
plt.show()
```

```python
# make predictions on the test set
y_pred = slr.predict(X_test)
```

```python
# Prediction of Test and Training set result
y_pred_slr = slr.predict(X_test)
```

```python
# Actual value and the predicted value
slr_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': y_pred_slr})
print(slr_diff)
```

```python
# Plot comparing the real and predicted values
plt.plot(y_test.values, color = 'red', label = 'Real data')
plt.plot(y_pred_slr, color = 'blue', label = 'Predicted data')
plt.title('Prediction')
plt.legend()
plt.show()
```
